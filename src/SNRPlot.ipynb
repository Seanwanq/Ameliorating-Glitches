```python
#%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = np.load("../Mock_Data/dataset_all_H1_bootstrap_O3a.npy")

# Create a new dataset with 'Class' set to 0
data_class_0 = data.copy()
data_class_0[:, -1] = 0

# Create a new dataset with 'Class' set to 1
data_class_1 = data.copy()
data_class_1[:, -1] = 1

pd.DataFrame(data_class_0)
```

```python
# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate the values
x_vals = data_class_1[:,0]
y_vals = data_class_1[:,2]
z_vals = data_class_1[:,3]

# Plot the values
ax.scatter(x_vals, y_vals, z_vals, c = 'b', marker='o')
ax.set_xlabel('SNR')
ax.set_ylabel('M1')
ax.set_zlabel('M2')

plt.show()
```
